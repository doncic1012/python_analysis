```python
import numpy as np
students = np.array([
    [67, 94, 65],
    [78, 92, 68],
    [59, 92, 87]
    ])
students
```

```python
import pandas as pd

pd.DataFrame(data=students,
            index=['stu1','stu2','stu3'],
            columns=['國文','英文','數學']
            )
```

```python
import random
def get_name(nums:int=10)->list[str]:
    with open('names.txt', encoding='utf-8') as file:
        content = file.read()
        names = content.split("\n")
        randam_names = random.sample(names, nums)
        
    return randam_names
get_name(5)
```

```python
student_nums = 50
data = np.random.randint(50, high=101, size=(student_nums,5))
students = pd.DataFrame(data,
            columns=['國文', '英文', '數學', '地理', '探討'],
            index=get_name(student_nums))
students
```

```python
students.to_csv('student.csv')
print("csv存檔成功")
```
